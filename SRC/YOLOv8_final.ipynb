```python
# load packages
import fiftyone as fo
import fiftyone.zoo as foz
from fiftyone import ViewField as F
import numpy as np
import os
from tqdm import tqdm
```

```python
# import the model
from ultralytics import YOLO

detection_model = YOLO("yolov8n.pt")
```

```python
import os

# path to the folder containing the images
images_folder = "/Users/treypallace/Desktop/DS4002 Project 2/Data/cville_roads"

# list all the files in the folder
image_files = [os.path.join(images_folder, file) for file in os.listdir(images_folder) if file.endswith(('.jpg', '.jpeg', '.png'))]

# perform object detection on each file
results = []
for image_file in image_files:
    result = detection_model(image_file)
    results.append(result)

# print or process the results as needed
for idx, res in enumerate(results):
    print(f"Results for {image_files[idx]}:")
    print(res)
    print("\n")
```
